```python
#load necessary packages
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

%matplotlib inline 
#formatting the graphs when they are show 
```

```python
#loading data with pandas. 
data=pd.read_csv('https://raw.githubusercontent.com/tiwariabhishekk10/Datasets/main/development_indicators.csv')
```

```python
print(data)
```

```python
#histogram of GDP per capita for the year 1960
#[0:162] to choose data from year 1960. to filter data for year 1960 use data[data.Year == 196
plt.hist(data['GDP per Capita'][0:162])
plt.xlabel('GDP Per Capita (USD)')
plt.ylabel('Count')
plt.title('GDP Per Capita in 1960')
plt.show()
```

```python
#summarise data 
gdp_mean = np.mean(data['GDP per Capita'][0:162])
gdp_med = np.median(data['GDP per Capita'][0:162])
gdp_std = np.std(data['GDP per Capita'][0:162])
```

```python
print(gdp_mean)
print(gdp_med)
print(gdp_std)
```

```python
#specify edge color of histogram box  
plt.hist(data['GDP per Capita'][0:162],ec='black')

#mean and median relative to data
#plt.axvline marks vertical line relative to x value
plt.axvline(gdp_mean,color='darkgray',label='Mean')
plt.axvline(gdp_med,color='darkgray',linestyle='-.',label='Mean')

#plt.axhline plot horizontal line relative to y value
#next two arguments specify endpoints of the line
xmin = gdp_mean - gdp_std
xmax = gdp_mean + gdp_std 
plt.hlines(40,xmin,xmax,color='darkgray',label='STD Range')

plt.xlabel('GDP Per Capita (USD)')
plt.ylabel('Count')
plt.title('GDP Per Capita in 1960')

#activates a legend for the data 
plt.legend()

plt.show()
```

```python
#Scatter plots to compare to attributes of the dataset
#alpha keyword - transparency of data point. useful when lot of overlapping data points
#s - size of points 
plt.scatter(data['GDP per Capita'][0:162],data['C02 Emissions per Person (Tonnes)'][0:162],alpha=0.5,s=15)

plt.xlabel('GDP per Capita')
plt.ylabel('CO2 Emission/Person')
plt.title('CO$_2$ vs GDP')

plt.show()
```

```python
#Color code the data points (countries) by the contitent

for i in range(0,162):
    if data['Continent'][i] == 'Asia':
        colo='green'
    if data['Continent'][i] == 'Americas':
        colo='blue'
    if data['Continent'][i] == 'Europe':
        colo='orange'
    if data['Continent'][i] == 'Europe':
        colo='red'
    if data['Continent'][i] == 'Ocenia':
        colo='yellow'
    plt.scatter(data['GDP per Capita'][i],data['C02 Emissions per Person (Tonnes)'][i],alpha=0.5,s=15,c=colo)    

plt.xlabel('GDP per Capita')
plt.ylabel('CO2 Emission/Person')
plt.title('CO$_2$ vs GDP')

plt.show()
        
```
